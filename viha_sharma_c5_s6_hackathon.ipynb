# HACKATHON - THE SHOPPING CART PROBLEM

## STEP 1:- CREATING CONSTANTS

```python
# making inventory 

chocolate_mousse = 15
red_velvet = 10
blueberry_cheesecake = 12
caramel_brownie = 14
apple_crumble = 5
bobo_special = 20


# cost per unit

cost_chocolate_mousse = 235
cost_red_velvet = 325
cost_blueberry_cheesecake = 300
cost_caramel_brownie = 210
cost_apple_crumble = 295
cost_bobo_special = 250

from datetime import date
```

```python
def add_to_basket():
    print('\033[1;3mʕ´•ᴥ•`ʔ WELCOME TO BOBO\'S BAKERY!\033[0m')
    print('ᴮᵒᵇᵒ ᵇᵃᵏᵉˢ ᵇᵉˢᵗ\n')
    print('                 \033[1;3m❀BAKERY MENU❀\n-----------------------------------------------')
    print('\033[1;36m%s %12s %11s %17s\033[0m' %('S.NO', 'ITEM','QTY','PRICE/SLICE(₹)'))
    print('-----------------------------------------------')
    print('%i %20s %7i %13.2f\n' %(1, 'Chocolate Mousse', chocolate_mousse, cost_chocolate_mousse ))
    print('%i %20s %7i %13.2f\n' %(2, 'Red Velvet Cake', red_velvet, cost_red_velvet))   
    print('%i %22s %5i %13.2f\n' %(3, 'Blueberry Cheesecake', blueberry_cheesecake, cost_blueberry_cheesecake))
    print('%i %20s %7i %13.2f\n' %(4, 'Caramel Brownie', caramel_brownie, cost_caramel_brownie)) 
    print('%i %20s %7i %13.2f\n' %(5, 'Apple Crumble', apple_crumble , cost_apple_crumble)) 
    print('%i %20s %7i %13.2f\n' %(5, 'Bobo Special', bobo_special , cost_bobo_special)) 
    
    
    qty_mousse = int(input('Enter the quantity of Chocolate Mousse: '))
    if qty_mousse > chocolate_mousse:
        print('Sorry, we have only', chocolate_mousse,'in stock!')
        qty_mousse = 0
    qty_redvel = int(input('Enter the quantity of Red Velvet Cake: '))
    if qty_redvel > red_velvet:
        print('Sorry, we have only', red_velvet,'in stock!')
        qty_redvel = 0
    qty_blueb = int(input('Enter the quantity of Blueberry Cheesecake: '))
    if qty_blueb > blueberry_cheesecake:
        print('Sorry, we have only', blueberry_cheesecake,'in stock!')
        qty_blueb = 0
    qty_caramel = int(input('Enter the quantity of Caramel Brownie: '))
    if qty_caramel > caramel_brownie:
        print('Sorry, we have only', caramel_brownie,'in stock!')
        qty_caramel = 0    
    qty_apple = int(input('Enter the quantity of Apple Crumble: '))
    if qty_apple > apple_crumble:
        print('Sorry, we have only', apple_crumble,'in stock!')
        qty_apple = 0    
    qty_bobo = int(input('Enter the quantity of Bobo Special ʕ♡˙ᴥ˙♡ʔ: '))
    if qty_bobo > bobo_special:
        print('Sorry, we have only', apple_crumble,'in stock!')
        qty_bobo = 0   
    return qty_mousse,qty_redvel,qty_blueb,qty_caramel,qty_apple, qty_bobo

    
def cal_total_item_cost(qty_mousse,qty_redvel,qty_blueb,qty_caramel,qty_apple, qty_bobo):
    total_cost = qty_mousse*cost_chocolate_mousse + qty_redvel*cost_red_velvet + qty_blueb*cost_blueberry_cheesecake + qty_caramel*cost_caramel_brownie + qty_apple*cost_apple_crumble + qty_bobo*cost_bobo_special
    return total_cost
    
def get_delivery_details(): 
    print()
    print('-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*')
    print('\033[1;3mʕ´•ᴥ•`ʔ☞ Bobo requires Customer Details!\033[0m ')
    print('-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*')
    name = input('Enter your Name: ')
    address = input('Enter your Address: ')
    dist = int(input('Distance from Bakery: '))
    return name, address, dist

def cal_delivery_charge(dist):
    if dist <=5:
        delivery_charge = 0
    else:
        delivery_charge = 100
    return delivery_charge 

def cal_final_bill(total_cost,delivery_charge):
    final_bill_amount = total_cost + delivery_charge
    return final_bill_amount

def generate_bill(qty_mousse,qty_redvel,qty_blueb,qty_caramel,qty_apple, qty_bobo, total_cost, final_bill_amount, name, address):
    print()
    print('-----------------------------------')    
    print('\033[1;3mʕ♡˙ᴥ˙♡ʔ YOUR ORDER HAS BEEN PLACED.\033[0m')
    print('-----------------------------------\n')
    print('                    \033[1;3m❀🍴YOUR BOBO BILL🍴❀\n-------------------------------------------------------')
    print('\033[1;36mDate:\033[0m', date.today(), '\n')
    print('\033[1;36m%s %12s %14s %10s %11s\033[0m' %('S.NO', 'ITEM','QTY','PRICE/SLICE', 'TOTAL COST'))
    print('-------------------------------------------------------')
    print('%i %20s %7i %13.2f %10.2f\n' %(1, 'Chocolate Mousse', qty_mousse, cost_chocolate_mousse, cost_chocolate_mousse*qty_mousse))
    print('%i %20s %7i %13.2f %10.2f\n' %(2, 'Red Velvet Cake', qty_redvel, cost_red_velvet, cost_red_velvet*qty_redvel))   
    print('%i %22s %5i %13.2f %10.2f\n' %(3, 'Blueberry Cheesecake', qty_blueb, cost_blueberry_cheesecake, cost_blueberry_cheesecake*qty_blueb))
    print('%i %20s %7i %13.2f %10.2f\n' %(4, 'Caramel Brownie', qty_caramel, cost_caramel_brownie,cost_caramel_brownie*qty_caramel )) 
    print('%i %20s %7i %13.2f %10.2f\n' %(5, 'Apple Crumble', qty_apple , cost_apple_crumble, cost_apple_crumble*qty_apple)) 
    print('%i %20s %7i %13.2f %10.2f\n' %(5, 'Bobo Special', qty_bobo , cost_bobo_special, cost_bobo_special*qty_bobo)) 
    
    print('\033[1;36mTOTAL COST:\033[0m','₹',total_cost)
    print()
    print('\033[1;36mTOTAL BILL (Total Cost + Delivery Charges):\033[0m', '₹', final_bill_amount)
    print()
    print('Bill made out to: ')
    print()
    print('\033[1;36mName:\033[0m', name, '\n')
    print('\033[1;36mAddress:\033[0m',address, '\n')

while True:
    qty_mousse,qty_redvel,qty_blueb,qty_caramel,qty_apple, qty_bobo = add_to_basket()
    total_cost = cal_total_item_cost(qty_mousse,qty_redvel,qty_blueb,qty_caramel,qty_apple, qty_bobo)
    chocolate_mousse = chocolate_mousse - qty_mousse
    red_velvet = red_velvet - qty_redvel
    blueberry_cheesecake = blueberry_cheesecake - qty_blueb
    caramel_brownie = caramel_brownie - qty_caramel
    apple_crumble = apple_crumble - qty_apple
    bobo_special = bobo_special - qty_bobo
    name, address, dist = get_delivery_details()
    delivery_charge = cal_delivery_charge(dist)
    final_bill_amount = cal_final_bill(total_cost,delivery_charge)
    generate_bill(qty_mousse,qty_redvel,qty_blueb,qty_caramel,qty_apple, qty_bobo, total_cost, final_bill_amount, name, address)
    x = input('Would you like to continue (y/n)? ')
    print('-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*')
    if x == 'n':
        print('\033[1;3mʕ´•ᴥ•`ʔ THANKS FOR SHOPPING AT BOBO\'S BAKERY!\033[0m')
        print('ᴮᵒᵇᵒ ᵇᵃᵏᵉˢ ᵇᵉˢᵗ\n')
        break
```
