## <font color='grey'> **ℙ𝕣𝕠𝕘𝕣𝕒𝕞 𝔽𝕦𝕟𝕔𝕥𝕚𝕠𝕟𝕒𝕝𝕚𝕥𝕪**:

&#10148; Program contains details of tablets along with their price

&#10148; Program allows user to order the tablets

&#10148; Program generates final bill

&#10148; Program takes user input for name, address and distance from store to delivery medicine.

```python
#Making inventory
paracetamol = 20
crocin = 15
zifi = 30
dolo = 40

#cost per unit
cost_paracetamol = 5.5
cost_crocin = 10
cost_zifi = 15.5
cost_dolo = 20

def add_to_basket():
    print('           AVAILABLE MEDICINES\n-----------------------------------------')
    print('\033[1;36m%s %10s %10s %11s\033[0m' %('S.NO', 'MED NAME','QUANTITY','PRICE/COST'))
    print('%i %14s %5i %10.2f' %(1, 'Paracetamol', paracetamol, cost_paracetamol))
    print('%i %13s %6i %10.2f' %(2, 'Crocin', crocin, cost_crocin))   
    print('%i %13s %6i %10.2f' %(3, 'Zifi', zifi, cost_zifi))
    
    qty_para = int(input('Enter the quantity of Paracetamol: '))
    qty_crocin = int(input('Enter the quantity of Crocin: '))
    qty_zifi = int(input('Enter the quantity of Zifi: '))
    return qty_para, qty_crocin,qty_zifi
    
def cal_total_item_cost(qty_para, qty_crocin,qty_zifi):
    total_cost = qty_para*cost_paracetamol + qty_crocin*cost_crocin + qty_zifi*cost_zifi
    return total_cost

def get_delivery_details():
    print('Enter your delivery details: ')
    name = input('Enter your name: ')
    address = input('Enter your address: ')
    dist = int(input('Distance from store: '))
    return name, address, dist

def cal_delivery_charge(dist):
    if dist <=5:
        delivery_charge = 0
    else:
        delivery_charge = 100
    return delivery_charge 

def cal_final_bill(total_cost,delivery_charge):
    final_bill_amount = total_cost + delivery_charge
    return final_bill_amount

def generate_bill(qty_para, qty_crocin,qty_zifi, total_cost, final_bill_amount, name, address):
    print('--------------')    
    print('YOUR DELIVERY BILL')
    print('--------------')
    print('\033[1;36m%s %10s %10s %11s %10s\033[0m\n' %('S.NO', 'MED NAME','QUANTITY','PRICE/SLICE', 'TOTAL COST'))
    print('%i %14s %5i %10.2f %10.2f\n' %(1, 'Paracetamol', qty_para, cost_paracetamol,qty_para*cost_paracetamol))
    print('%i %14s %5i %10.2f %10.2f\n' %(2, 'Crocin', qty_crocin, cost_crocin, qty_crocin*cost_crocin))   
    print('%i %14s %5i %10.2f %10.2f\n' %(3, 'Zifi', qty_zifi, cost_zifi, qty_zifi*cost_zifi))
    
    print('Total Cost of Items:', total_cost)
    print('TOTAL BILL : Total Cost of Items + Delivery Charges: ', final_bill_amount)
    print('Name: ', name)
    print('Address: ',address)

    
while True:
    qty_para, qty_crocin,qty_zifi = add_to_basket()
    total_cost = cal_total_item_cost(qty_para, qty_crocin,qty_zifi)
    print(total_cost)
    paracetamol = paracetamol - qty_para
    crocin = crocin - qty_crocin
    zifi = zifi - qty_zifi
    name, address, dist = get_delivery_details()
    delivery_charge = cal_delivery_charge(dist)
    final_bill_amount = cal_final_bill(total_cost,delivery_charge)
    generate_bill(qty_para, qty_crocin,qty_zifi, total_cost, final_bill_amount, name, address)
    x = input('Would you like to continue (y/n)? ')
    if x == 'n':
        print('Thank you for shopping with us.')
        break
        
```
